```python
import pandas as pd
import numpy as np

# 成交量数据 (万股) - 扩展到8天
volume = pd.DataFrame(
    {
        "stock_A": [1000, 1200, 1100, 1300, 1150, 1250, 1080, 1320],
        "stock_B": [800, 900, 850, 950, 880, 920, 840, 980],
        "stock_C": [1500, 1600, 1550, 1700, 1580, 1650, 1520, 1720],
    },
    index=["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7", "Day8"],
)

# 开盘价数据 - 扩展到8天
open_price = pd.DataFrame(
    {
        "stock_A": [10.0, 10.5, 10.2, 10.8, 10.4, 10.6, 10.3, 10.9],
        "stock_B": [8.0, 8.2, 8.1, 8.4, 8.2, 8.3, 8.0, 8.5],
        "stock_C": [12.0, 12.3, 12.1, 12.6, 12.2, 12.4, 12.0, 12.7],
    },
    index=["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7", "Day8"],
)

# 收盘价数据 - 扩展到8天
close = pd.DataFrame(
    {
        "stock_A": [10.5, 10.2, 10.8, 10.4, 10.7, 10.3, 10.9, 10.6],
        "stock_B": [8.2, 8.1, 8.4, 8.2, 8.3, 8.0, 8.5, 8.3],
        "stock_C": [12.3, 15.1, 12.6, 18.2, 12.5, 10.0, 12.7, 12.4],
    },
    index=["Day1", "Day2", "Day3", "Day4", "Day5", "Day6", "Day7", "Day8"],
)
```
